```python
str = "98912345678"
result = re.findall("\d{11}",str)
print(result)

result2= re.findall("\D{11}",str)
print(result2)

result3= re.findall("\d[11]",str)
print(result3)

result4= re.findall("\D[11]",str)
print(result4)
```
